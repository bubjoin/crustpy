```python
# 참고로, 재귀는 중요한 개념
# 자료구조에서도 높은 비율, 학부 알고리즘에도 상당 부분
# '쉽게 배우는 자료구조 with 파이썬', 문병로 교수님 저서 참고
# 재귀를 모르면 자료구조와 알고리즘을 공부할 수 없음

# 큰 문제와 그보다 더 작고 단순한 문제의 관계를 이용하여 문제 해결
# 경계 조건, 재귀 호출, 관계 표현으로 구성됨

# 가장 단순한 예는 팩토리얼

def fac(n):
  if n==1:
    return 1
  return n * fac(n-1)

print(fac(1))
print(fac(2))
print(fac(3))
print(fac(4))
print(fac(5))
print(fac(6))
```

```python
# 작동은 하는 코드 VS 더 나은 코드

# 문제를 보고
# 대충 이렇게 하면 되겠지

# 피보나치 수열의 값을 구하는 함수
# 일단 논리적이고 답도 맞음, 그런데...
def fib(n):
  if (n==1) or (n==2):
    return 1
  else:
    return fib(n-1) + fib(n-2)

n = 10
answer = fib(n)
print(answer)

n = 20
answer = fib(n)
print(answer)

n = 30
answer = fib(n)
print(answer)
```

```python
# '쉽게 배우는 자료구조 with 파이썬', 문병로 교수님 저서 참고
# 자료구조를 공부해야하는 이유
# 피보나치 수열

calc_num = 0
def fib(n):
  global calc_num
  calc_num += 1
  if (n==1) or (n==2):
    return 1
  else:
    return fib(n-1) + fib(n-2)

n = 40
print(f'fib({n}): {fib(n):d}, calc_num: {calc_num}')
```

```python
# '쉽게 배우는 자료구조 with 파이썬', 문병로 교수님 저서 참고
# 더 빠르게 구하는 방법
# (느렸던 이유는 이미 계산된 것을 또 계산해서)

calc_num=0
def fib_fast(n):
  f = [0] * (n+1)
  f[1], f[2] = 1, 1
  global calc_num
  calc_num = 2
  for i in range(3, n+1):
    f[i] = f[i-1] + f[i-2]
    calc_num += 1
  return f[n]

n = 40
print(f'fib_fast({n}): {fib_fast(n):d}, calc_num: {calc_num}') 
```

```python
# 우리반애들
# 스포츠키드

class Kid:
  kids = 0
  def __init__(self, name):
    self.name = name
    Kid.kids = Kid.kids + 1


class SportsKid(Kid):
  def __init__(self, name):
    super().__init__(name)
    speed = 0

  def set_speed(self, speed):
    self.speed = speed

  def get_speed(self):
    return self.speed
```

```python
# 키드를 상속받은 스포츠키드 클래스로 오브젝트를 생성

friends = []

for i in range(0, 4):
  kid = SportsKid(f'스포츠꿈나무 {i}')
  friends.append(kid)

for kid_obj in friends:
  print(kid_obj)

print(f'스포츠 키즈는 총 {SportsKid.kids}명 입니다.')
```

```python
from random import randint

for kid in friends:
  kid.set_speed(randint(40,100))

for kid in friends:
  print(f'{kid.name} : 스피드는 {kid.get_speed():>3d}')
```

```python
# 스피드 경기

fast_kid = SportsKid('기준')
fast_kid.set_speed(0)

for kid in friends:
  if kid.get_speed() > fast_kid.get_speed():
    fast_kid = kid

print(f'가장 빠른 친구는 {fast_kid.name}이고, 스피드는 {fast_kid.get_speed()} 입니다.')
```

```python
import os
import datetime

def list_by_name():
  print(os.getcwd()) # 현재 작업 중인 디렉토리를 보여줌
  files = os.listdir() # 디렉토리에 있는 파일들을 리스트로 반환
  files = sorted(files, key = lambda x : x)
  for file in files:
    seconds = os.path.getmtime(file)
    modified_date = datetime.datetime.utcfromtimestamp(seconds)
    print(f'{modified_date}   {file}')

def list_by_time():
  print(os.getcwd())
  files = os.listdir()
  files = sorted( files, key = lambda x : os.path.getmtime(x))
  for file in files:
    seconds = os.path.getmtime(file)
    modified_date = datetime.datetime.utcfromtimestamp(seconds)
    print(f'{modified_date}   {file}')

list_by_time()
```
